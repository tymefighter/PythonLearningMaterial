```
from IPython.display import clear_output
```

```
def takeInput():
    x1 = int(input("please enter first dot row number: "))
    y1 = int(input("please enter first dot col number: "))
    x2 = int(input("please enter second dot row number: "))
    y2 = int(input("please enter second dot col number: "))

    return (x1,y1),(x2,y2)
```

```
class DotDash:

    def __init__(self, n):
        self.n = n
        self.TotalScore = (n - 1) * (n - 1)

        self.horz = [[False for j in range(n - 1)] for i in range(n)]
        self.vert = [[False for j in range(n)] for i in range(n - 1)]
        self.own = [[' ' for j in range(n)] for i in range(n)]

        self.playerNo = 0
        self.pScore = [0, 0]

    def checkValid(self, p1, p2):

        x1 = p1[0]
        y1 = p1[1]
        x2 = p2[0]
        y2 = p2[1]
        
        if  x1 >= self.n or x1 < 0 or y1 >= self.n or y1 < 0:
            return False
            
        elif x2 >= self.n or x2 < 0 or y2 >= self.n or y2 < 0:
            return False
        
        elif x1 == x2 and abs(y1 - y2) == 1 and self.horz[x1][min(y1,y2)] == False:
            return True
            
        elif y1 == y2 and abs(x1 - x2) == 1 and self.vert[min(x1,x2)][y1] == False:
            return True
            
        else:
            return False
    
    def update(self, p1, p2):
	
        x1 = p1[0]
        y1 = p1[1]
        x2 = p2[0]
        y2 = p2[1]
        
        Flag = True
        
        if x1 == x2:
            x = x1
            y = min(y1,y2)
            self.horz[x][y] = True
            
            if x>0 and y<(self.n-1) and  self.horz[x-1][y] and self.vert[x-1][y] and self.vert[x-1][y+1]:
                Flag = False
                self.pScore[playerNo] += 1
                self.own[x-1][y] = str(self.playerNo)
                    
            if x<(self.n-1) and y<(self.n-1) and  self.horz[x+1][y] and self.vert[x][y] and self.vert[x][y+1]:
                Flag = False
                self.pScore[self.playerNo] += 1
                self.own[x][y] = str(self.playerNo)
        else:
            x = min(x1,x2)
            y = y1
            self.vert[x][y] = True
            
            if x<(self.n-1) and y<(self.n-1) and self.horz[x][y] and self.horz[x+1][y] and self.vert[x][y+1]:
                Flag = False
                self.pScore[playerNo] += 1
                self.own[x][y] = str(self.playerNo)
                
            if y>0 and x<(self.n-1) and self.vert[x][y-1] and self.horz[x][y-1] and self.horz[x+1][y-1]:
                Flag = False
                self.pScore[playerNo] += 1
                self.own[x][y-1] = str(self.playerNo)
                
        if Flag:
            self.playerNo = 1-self.playerNo

    def display(self):
	
        s = "   "
        
        for j in range(2 * self.n - 1):
            
            if j % 2 == 0:
                s += str(j // 2)
            else:
                s += ' '
        s += "\n\n"
        
        for i in range(2 * self.n - 1):
            
            
            if i % 2 == 0:
                
                s += str(i // 2) + "  "	
                    
                for j in range(2 * self.n - 1):
                    
                    if j % 2 == 0:
                        s += '$'
                    elif self.horz[i // 2][j // 2]:
                        s += '-'
                    else:
                        s += ' '
            else:
            
                s += "   "
                
                for j in range(2 * self.n - 1):
                    
                    if j % 2 != 0:
                        s += self.own[i // 2][j // 2]
                    elif self.vert[i // 2][j // 2]:
                        s += '|'
                    else:
                        s += ' '
            
            s += "\n"
            
        
        clear_output()
        print(s)
 
    def isGameOver(self):

        """
            - if game is not over, then it would return -1
            - if player 0 won, return 0
            - if player 1 won, return 1
            - if scores are equal, return 2 to indicate draw
        """

        if self.pScore[0] + self.pScore[1] != self.TotalScore:
            return -1
        elif self.pScore[0] > self.pScore[1]:
            return 0
        elif self.pScore[1] > self.pScore[0]:
            return 1
        else:
            return 2
    
    def play(self):

        self.playerNo = 0
        won = -1

        while won == -1:
            
            self.display()
            p1, p2 = takeInput()
            
            if not self.checkValid(p1, p2):
                continue
            
            self.update(p1, p2)
            won = self.isGameOver()
        
        
        display()
        if won == 0:
            print("Player 0 won")
        elif won == 1:
            print("Player 1 won")
        else:
            print("Match Draw")
```
